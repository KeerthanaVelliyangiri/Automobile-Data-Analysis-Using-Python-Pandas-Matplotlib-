```python
import matplotlib.pyplot as plt
```

```python
toporgin=df.groupby('Origin').count()
print(toporgin)
```

```python
toporgin.plot.bar(legend=False)
plt.title('Origin Distribution')
plt.show()
```

```python
asia=df[df['Origin']=="Asia"].groupby('Make').count()
print(asia)
```

```python
asia['Model'].plot.bar(color='green')
plt.title('Number of Asian Cars by Make')
plt.show()
```

```python
europe=df[df['Origin']=="Europe"].groupby('Make').count()
print(europe)
```

```python
europe['Model'].plot.bar(color='purple')
plt.title('Number of Europe Cars by Make')
plt.show()
```

```python
usa=df[df['Origin']=="USA"].groupby('Make').count()
print(usa)
```

```python
usa['Model'].plot.bar(color='blue')
plt.title('Number of USA Cars by Make')
plt.show()
```

```python
cartype=df['Type'].nunique()
print(cartype)
```

```python
cartypes=df['Type'].value_counts()
print(cartypes)
```

```python
cartypes.plot.pie()
plt.title('Car Types Distribution')
plt.show()
```

```python
topcar = df.sort_values('MSRP', ascending=False).head(5)[['Make', 'Model', 'MSRP', 'Horsepower']]
print(topcar)
```

```python
df['MSRP']=df['MSRP'].astype(str).str.replace('$','',regex=False).str.replace(',','',regex=False).astype(float)
df['MSRP']=pd.to_numeric(df['MSRP'])
brandavgprices=df.groupby('Make')['MSRP'].mean()
print(brandavgprices)
```

```python
brandavgprices = df.groupby('Make')['MSRP'].mean()
print(brandavgprices)
```

```python
brandavgprices.nlargest(15).plot(kind='bar')
plt.title('Top 15 Most Expensive Car Makes (Average MSRP)')
plt.ylabel('Average Price')
plt.show()
```

```python
top5 = brandavgprices.nlargest(5)
print(top5)
```

```python
df['Invoice']=df['Invoice'].astype(str).str.replace('$','',regex=False).str.replace(',','',regex=False).astype(float)
df['Invoice']=pd.to_numeric(df['Invoice'])
```

```python
df['Pricediff'] = df['MSRP'] - df['Invoice']
print(df['Pricediff'])
```
